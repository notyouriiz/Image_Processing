# Image Color Representations  

## 1. RGB Image  
- **Definition:** RGB stands for **Red, Green, and Blue**.  
- **How it works:** Each pixel is represented by three values (R, G, B), each ranging from 0–255.  
- **Result:** By combining these three channels, millions of colors can be represented.  
- **Example:**  
  - Pure red → (255, 0, 0)  
  - Pure green → (0, 255, 0)  
  - Pure blue → (0, 0, 255)  

---

## 2. Grayscale Image  
- **Definition:** A grayscale image only contains **intensity information** instead of full color.  
- **How it works:** The RGB values are converted into a single value (0–255), where:  
  - 0 → black  
  - 255 → white  
  - values in between → shades of gray  
- **Result:** A 2D image with brightness variations but no color.  

---

## 3. Black and White (Binary) Image  
- **Definition:** A black and white image (binary image) has only two possible pixel values:  
  - 0 → black  
  - 255 → white  
- **How it works:** Created by applying a **threshold** to the grayscale image:  
  - If pixel ≥ threshold → White (255)  
  - If pixel < threshold → Black (0)  
- **Result:** A high-contrast image useful for object detection, segmentation, or text extraction.  

---

## ✅ Summary  
- **RGB** → Full color image with 3 channels (R, G, B).  
- **Grayscale** → Single channel showing brightness (intensity).  
- **Black & White** → Binary image after thresholding (only black or white).  


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("assets/Image.jpg")
```

```python
# Read Grayscale Image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Read Black and White image
thresh, BW_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)

# Print Image
# Show results inline
plt.figure(figsize=(12,5))

# Original (convert BGR to RGB for correct colors)
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

# Grayscale
plt.subplot(1,3,2)
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# Black & White
plt.subplot(1,3,3)
plt.imshow(BW_image, cmap="gray")
plt.title("Black & White")
plt.axis("off")

plt.show()
```
