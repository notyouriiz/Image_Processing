# Image Affine Transformation Example (OpenCV + Matplotlib)

In this notebook, we demonstrate how to apply **Affine transformations** to an image using OpenCV:  
- **Horizontal Shear** (slant left/right)  
- **Vertical Shear** (slant up/down)  

---

## 1. Import Libraries
We need:
- `cv2` (OpenCV) → for reading and transforming the image
- `numpy` → for transformation matrices
- `matplotlib.pyplot` → for displaying images

## 2. Load the Image

We load the image from `assets/image.png`.  
Since OpenCV loads images in **BGR format**, we convert it to **RGB** for correct display in Matplotlib.

## 3. Define Affine Transformation Matrices

The general form of an **Affine Transformation Matrix** is:

Horizontal shear:

$$
M_h = \begin{bmatrix}
1 & s_x & 0 \\
0 & 1 & 0
\end{bmatrix}
$$

Vertical shear:

$$
M_v = \begin{bmatrix}
1 & 0 & 0 \\
s_y & 1 & 0
\end{bmatrix}
$$


- \(s_x\) = horizontal shear factor  
- \(s_y\) = vertical shear factor  
- \(t_x\), \(t_y\) = translation (shift in x and y)


- For **horizontal shear** → add shear to the first row.  
- For **vertical shear** → add shear to the second row.

## 4. Apply Affine Transformation

We use `cv2.warpAffine()` with our matrices.

## 5. Show Images Side by Side


## Import Libraries and Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image (convert BGR → RGB for display)
image = cv2.imread("assets/Image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Image Affined

```python
# Image size
h, w = image.shape[:2]

# Shear factors
shear_factor = 0.4

# Horizontal shear matrix
M_horizontal = np.float32([
    [1, shear_factor, 0],
    [0, 1, 0]
])

# Vertical shear matrix
M_vertical = np.float32([
    [1, 0, 0],
    [shear_factor, 1, 0]
])

# Apply horizontal and vertical affine transforms
shear_horizontal = cv2.warpAffine(image, M_horizontal, (w, h))
shear_vertical = cv2.warpAffine(image, M_vertical, (w, h))

# Display original, horizontal shear, vertical shear
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(shear_horizontal)
axes[1].set_title("Horizontal Shear")
axes[1].axis("off")

axes[2].imshow(shear_vertical)
axes[2].set_title("Vertical Shear")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
