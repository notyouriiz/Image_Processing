# Image Rotation Example (OpenCV + Matplotlib)

In this notebook, we demonstrate how to **rotate an image** using OpenCV and display the results **side by side** with Matplotlib.

---

## 1. Import Libraries
We need:
- `cv2` (OpenCV) → for reading and rotating the image
- `matplotlib.pyplot` → for displaying images in the notebook

## 2. Load the Image

We load the image from `assets/image.png`.  
Since OpenCV loads images in **BGR format**, we convert it to **RGB** for correct display in Matplotlib.

## 3. Apply Rotation

OpenCV provides:
- `cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)`
- `cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)`
- `cv2.rotate(image, cv2.ROTATE_180)`

## 4. Show Images Side by Side

We use Matplotlib’s `subplots` to arrange **Original, 90° CW, 90° CCW, 180°** in one row.


## Import Libraries

```python
import cv2
import matplotlib.pyplot as plt

# Load image (convert BGR → RGB for display)
image = cv2.imread("assets/Image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Image Rotation

```python
# Rotate the image
rotate_90_cw = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotate_90_ccw = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image, cv2.ROTATE_180)

# Display side by side
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title("Original")

axes[1].imshow(rotate_90_cw)
axes[1].set_title("90° Clockwise")

axes[2].imshow(rotate_90_ccw)
axes[2].set_title("-90°")

axes[3].imshow(rotate_180)
axes[3].set_title("180°")

# Remove axes for cleaner view
for ax in axes:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
