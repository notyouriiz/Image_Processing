# Image Scaling Example (OpenCV + Matplotlib)

In this notebook, we demonstrate how to **scale (resize) an image** using OpenCV and display the results **side by side** with Matplotlib.

---

## 1. Import Libraries
We need:
- `cv2` (OpenCV) → for reading and resizing the image
- `matplotlib.pyplot` → for displaying images in the notebook

## 2. Load the Image

We load the image from `assets/image.png`.  
Since OpenCV loads images in **BGR format**, we convert it to **RGB** for correct display in Matplotlib.

## 3. Apply Scaling (Resize)

OpenCV’s `cv2.resize()` allows scaling:
- By specific **scaling factors** (`fx`, `fy`)
- Or to a specific **width × height**

Here we scale:
- **0.5× (Downscale)** → shrink image  
- **1.5× (Upscale)** → enlarge image  
- **2× (Upscale)** → enlarge more

## 4. Show Images Side by Side

We use Matplotlib’s `subplots` to arrange **Original, 0.5×, 1.5×, 2×** in one row.


## Import Libraries

```python
import cv2
import matplotlib.pyplot as plt

# Load image (convert BGR → RGB for display)
image = cv2.imread("assets/Image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Image Scaling

```python
# Scale by factors (fx = width scale, fy = height scale)
scale_half = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
scale_1_5x = cv2.resize(image, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
scale_double = cv2.resize(image, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LINEAR)

# Display side by side
fig, axes = plt.subplots(1, 4, figsize=(18, 5))

axes[0].imshow(image)
axes[0].set_title("Original")

axes[1].imshow(scale_half)
axes[1].set_title("0.5× (Downscale)")

axes[2].imshow(scale_1_5x)
axes[2].set_title("1.5× (Upscale)")

axes[3].imshow(scale_double)
axes[3].set_title("2× (Upscale)")

# Remove axes for cleaner view
for ax in axes:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
