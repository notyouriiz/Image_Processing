# Image Flipping Example (OpenCV + Matplotlib)

In this notebook, we demonstrate how to **flip an image** using OpenCV and display the results **side by side** with Matplotlib.

---

## 1. Import Libraries
We need:
- `cv2` (OpenCV) → for reading and flipping the image
- `matplotlib.pyplot` → for displaying images in the notebook

## 2. Load the Image

We load the image from `assets/image.png`.  
Since OpenCV loads images in **BGR format**, we convert it to **RGB** for correct display in Matplotlib.

## 3. Apply Flipping

OpenCV’s `cv2.flip(image, flipCode)` allows flipping:
- `flipCode = 0` → Vertical (upside down)
- `flipCode = 1` → Horizontal (mirror)
- `flipCode = -1` → Both (vertical + horizontal)

## 4. Show Images Side by Side

We use Matplotlib’s `subplots` to arrange **Original, Vertical Flip, Horizontal Flip, Both Flip** in one row.

## Import Libraries

```python
import cv2
import matplotlib.pyplot as plt

# Load image (BGR -> RGB for matplotlib display)
image = cv2.imread("assets/image.png")
```

## Image Flipping

```python
# Load image (BGR -> RGB for matplotlib display)
image = cv2.imread("assets/image1.jpeg")

# Flip vertically (0), horizontally (1), and both (-1)
flip_vertical = cv2.flip(image, 0)
flip_horizontal = cv2.flip(image, 1)
flip_both = cv2.flip(image, -1)

# Display side by side
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title("Original")

axes[1].imshow(flip_vertical)
axes[1].set_title("Vertical Flip")

axes[2].imshow(flip_horizontal)
axes[2].set_title("Horizontal Flip")

axes[3].imshow(flip_both)
axes[3].set_title("Both Flip")

# Remove axes
for ax in axes:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
