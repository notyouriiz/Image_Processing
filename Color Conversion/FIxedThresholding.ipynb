# ðŸ”² Fixed Thresholding in Image Processing  

**Fixed Thresholding** is the simplest form of thresholding, where a single constant threshold value (**T**) is chosen manually. Every pixel in the image is compared against this threshold to decide whether it becomes **black (0)** or **white (255)**.  

---

## ðŸ”¹ Formula  

\[
BW(x,y) = 
\begin{cases} 
255 & \text{if } I(x,y) \geq T \\ 
0 & \text{if } I(x,y) < T 
\end{cases}
\]

- \( I(x,y) \): pixel intensity at position \((x,y)\).  
- \( T \): manually chosen threshold value (fixed for the entire image).  
- Output: a **binary image** with pixel values only **0** or **255**.  

---

## ðŸ”¹ Steps  
1. Convert the original image to **Grayscale** (if RGB).  
2. Select a **threshold value T** (commonly between 100â€“150).  
3. Apply the threshold rule to every pixel.  
   - If pixel â‰¥ T â†’ White (255).  
   - If pixel < T â†’ Black (0).  

---

## ðŸ”¹ Why Use Fixed Thresholding?  
âœ… Very **fast and simple** to apply.  
âœ… Good for **high-contrast images** with uniform lighting.  
âš ï¸ Not ideal for images with **shadows or varying brightness** (use Adaptive Thresholding instead).  

---

## ðŸ”¹ Example  
- **Original RGB** â†’ Full color image.  
- **Grayscale** â†’ Shades from 0â€“255.  
- **Fixed Thresholding (T = 127)** â†’ Image converted into **strict black & white** with no shades.  


```python
import cv2
import matplotlib.pyplot as plt

# --- Read Image ---
image = cv2.imread("assets/Image.jpg")  # replace with your file path
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# --- Apply Fixed Thresholding ---
T = 127  # fixed threshold value
_, bw_image = cv2.threshold(gray_image, T, 255, cv2.THRESH_BINARY)

# --- Visualization ---
plt.figure(figsize=(12,4))

# Original (convert BGR to RGB for correct colors)
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original RGB")
plt.axis("off")

# Grayscale
plt.subplot(1,3,2)
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# Black & White (Fixed Thresholding)
plt.subplot(1,3,3)
plt.imshow(bw_image, cmap="gray")
plt.title(f"Fixed Thresholding (T={T})")
plt.axis("off")

plt.tight_layout()
plt.show()

# --- Print some pixel values for verification ---
print("Sample grayscale pixel:", gray_image[50, 50])
print("After thresholding:", bw_image[50, 50])
```
