# RGB (Red, Green, Blue) Color Representation  

A digital color image is usually represented in the **RGB color model**.  
Each pixel is defined by three components:  

- **R** â†’ Red intensity (0â€“255)  
- **G** â†’ Green intensity (0â€“255)  
- **B** â†’ Blue intensity (0â€“255)  

By combining these three values, millions of colors can be represented.  

---

## ðŸ”¹ RGB Pixel Example  

- Pure **Red**: (255, 0, 0)  
- Pure **Green**: (0, 255, 0)  
- Pure **Blue**: (0, 0, 255)  
- Pure **White**: (255, 255, 255)  
- Pure **Black**: (0, 0, 0)  

---

## ðŸ”¹ Why Use RGB?  

- Standard format for most digital images and cameras.  
- Easy to manipulate individual color channels.  
- Basis for many conversions (e.g., to **Grayscale** or **HSV**).  

---

## ðŸ”¹ Example Output  
- **Original RGB**: Full color image with three channels (R, G, B).  
- **Red Channel**: Image formed using only red intensity values.  
- **Green Channel**: Image formed using only green intensity values.  
- **Blue Channel**: Image formed using only blue intensity values.  


## Import Libraries and Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an RGB image
img = cv2.imread("assets/Image.jpg")   # BGR format by default
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
```

## Conversion to RGB

```python
# --- Method 1: Built-in OpenCV Split ---
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R_builtin, G_builtin, B_builtin = cv2.split(img_rgb)

# --- Method 2: Manual Channel Extraction ---
R_manual = img_rgb[:, :, 0]
G_manual = img_rgb[:, :, 1]
B_manual = img_rgb[:, :, 2]

# --- Visualization ---
fig, ax = plt.subplots(2, 4, figsize=(14, 7))

# Original Image
ax[0,0].imshow(img_rgb)
ax[0,0].set_title("Original RGB")
ax[0,0].axis("off")

# Built-in Channels
ax[0,1].imshow(R_builtin, cmap="Reds")
ax[0,1].set_title("Red (Built-in)")
ax[0,1].axis("off")

ax[0,2].imshow(G_builtin, cmap="Greens")
ax[0,2].set_title("Green (Built-in)")
ax[0,2].axis("off")

ax[0,3].imshow(B_builtin, cmap="Blues")
ax[0,3].set_title("Blue (Built-in)")
ax[0,3].axis("off")

# Manual Channels
ax[1,0].imshow(img_rgb)
ax[1,0].set_title("Original RGB")
ax[1,0].axis("off")

ax[1,1].imshow(R_manual, cmap="Reds")
ax[1,1].set_title("Red (Manual)")
ax[1,1].axis("off")

ax[1,2].imshow(G_manual, cmap="Greens")
ax[1,2].set_title("Green (Manual)")
ax[1,2].axis("off")

ax[1,3].imshow(B_manual, cmap="Blues")
ax[1,3].set_title("Blue (Manual)")
ax[1,3].axis("off")

plt.tight_layout()
plt.show()
```
