# RGB/Grayscale to Black and White Conversion  

After converting an image to **grayscale**, each pixel has an intensity value between **0 (black)** and **255 (white)**.  
To create a **Black and White (Binary)** image, we apply a process called **thresholding**.  

---

## ðŸ”¹ Conversion Formula  

Given a **threshold value (T)**, each pixel is classified as either black or white:  

\[
BW(x,y) = 
\begin{cases} 
255 & \text{if } Gray(x,y) \geq T \\
0 & \text{if } Gray(x,y) < T 
\end{cases}
\]

- **T** = threshold (commonly between 100â€“150, depending on brightness).  
- Pixel values **greater than or equal to T** become white (255).  
- Pixel values **less than T** become black (0).  

---

## ðŸ”¹ Why Convert to Black and White?  

- **Simplifies images** into only two colors (black & white).  
- Useful for **object detection, OCR (text recognition), and segmentation**.  
- Removes unnecessary grayscale details, focusing only on contrast.  

---

## ðŸ”¹ Example Output  
- **Original RGB**: Full color image.  
- **Grayscale**: Intensity values from 0â€“255.  
- **Black & White**: Binary image (only black or white) after thresholding.  


## Import Libraries and Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an RGB image
img = cv2.imread("assets/Image.jpg")   # BGR format by default
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
```

## Conversion to Black & White

```python
# Convert to Grayscale first
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# --- Method 1: Built-in OpenCV Threshold ---
_, bw_builtin = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)

# --- Method 2: Manual Formula ---
# BW(x,y) = 255 if Gray >= T else 0
T = 150
bw_manual = np.where(gray_image >= T, 255, 0).astype(np.uint8)

# --- Visualization ---
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original RGB")
ax[0].axis("off")

ax[1].imshow(bw_builtin, cmap="gray")
ax[1].set_title("Black & White (Built-in)")
ax[1].axis("off")

ax[2].imshow(bw_manual, cmap="gray")
ax[2].set_title("Black & White (Manual Formula)")
ax[2].axis("off")

plt.tight_layout()
plt.show()
```
